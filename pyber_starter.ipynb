```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_data_to_load = pd.read_csv("data/city_data.csv")
ride_data_to_load = pd.read_csv("data/ride_data.csv")

# Combine the data into a single dataset

cityride_merged_df = pd.merge(city_data_to_load,ride_data_to_load ,how = "left", on ="city")



# Display the data table for preview
cityride_merged_df
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

average_fare_per_city = cityride_merged_df.groupby("city")["fare"].mean()

# Build the scatter plots for each city types
#Have to find the total number of rides per each city and average fair amongst them

#Total_rides
total_rides_per_city =cityride_merged_df.groupby("city")["ride_id"].count()
# total_rides_per_city



#Find the total number of drivers per city.

total_drivers_per_city = cityride_merged_df.groupby("city")["driver_count"].count()
#total_drivers_per_city

city_totals = city_data_to_load.set_index('city')['type']


city_totals.value_counts()

pyber_df = pd.DataFrame({"Average Fare per City":average_fare_per_city,
                 "Number of Rides": total_rides_per_city,
                 "Number of Drivers": total_drivers_per_city,
                 "City Type": city_totals})
pyber_df.head()


#Average
#Urban
Urban = pyber_df.loc[pyber_df["City Type"]== "Urban"]

#suburban
Suburban = pyber_df.loc[pyber_df["City Type"]== "Suburban"]

#rural
Rural = pyber_df.loc[pyber_df["City Type"]== "Rural"]

#Now convert the above data into scatter plots
#Urban
plt.scatter(Urban["Number of Rides"], Urban["Average Fare per City"], color = "pink", edgecolors="blue", s = Urban["Number of Drivers"]*10, label = "Urban", alpha = 0.7, linewidth = 1.5)

#Suburban
plt.scatter(Suburban["Number of Rides"], Suburban["Average Fare per City"], color = "yellow", edgecolors ="red", s = Suburban["Number of Drivers"]*10, label = "Suburban", alpha = 0.7, linewidth = 1.5)

#Rural
plt.scatter(Rural["Number of Rides"], Rural["Average Fare per City"], color = "grey", edgecolors = "brown", s = Rural["Number of Drivers"]*10, label = "Rural", alpha = 0.7, linewidth = 1.5)



# Incorporate the other graph properties
plt.title("Pyber Data 2018")
plt.xlabel("Total Number of Rides")
plt.ylabel("Average Fare Per City")

# Create a legend
plt.legend(loc= "upper right")

# # Incorporate a text label regarding circle size
# plt.text(0.5, 0.5, verticalalignment='upper right',horizontalalignment= 'Circle size is relevant to how many drivers are in that area per city' , fontsize=10)

plt.text(0.5, 0.5, "Circle size is relevant to how many drivers are in that area per city")
#plt.text("Circle size is relevant to how many drivers are in that area per city")
# Save Figure
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
# Calculate Ride Percents

#Sum of fares by city type.

city_fare_total = cityride_merged_df.groupby(["type"])["fare"].sum()

# Build Pie Chart

city_fare_total.index
explode = (0,0, 0.2)
colors = ["pink", "grey", "orange"]
plt.axis("equal")



plt.pie(city_fare_total, labels = city_fare_total.index, autopct = "%1.2f%%", colors = colors, explode = explode, shadow = True, startangle = 170)
# Build Pie Chart

plt.legend(loc = "lower right")
# Save Figure

plt.title("Total Fares By City Type")
#how 

# Show Figure
plt.show()
```
